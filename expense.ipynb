```python
import pdfplumber
import pandas as pd

pdf_path = "154.pdf"
pdf_password = "904302"  # Replace with your actual password

all_data = []

with pdfplumber.open(pdf_path, password=pdf_password) as pdf:
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            df = pd.DataFrame(table[1:], columns=table[0])
            all_data.append(df)

# Combine all pages
df = pd.concat(all_data, ignore_index=True)

# Save to clean CSV
df.to_csv("cleaned_mpesa.csv", index=False)

print("Cleaned CSV saved as cleaned_mpesa.csv")
```

```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Home2020@!",  # change this!
    database="fraud_detection"
)
cursor = conn.cursor()
df = pd.read_csv("cleaned_mpesa.csv")
print(df.columns.tolist())

# Check for 'nan' values in the entire DataFrame
print("\nNumber of 'nan' values in each column:")
print(df.apply(lambda col: col.astype(str).str.contains('nan').sum()))



 #Option 2: Replace 'nan' with a default value (e.g., 0 for numeric, '' for string)
df ['Paid In'] = df['Paid In'].replace('nan', 0)
df ['Withdrawn'] = df['Withdrawn'].replace('nan', 0)
df ['Balance'] = df['Balance'].replace('nan', 0)
df ['Details'] = df['Details'].replace('nan', '')
# ... and so on for other relevant columns

for _, row in df.iterrows():
    sql = """
        INSERT INTO transactions
        (receipt_no, completion_time, details, transaction_status, paid_in, withdrawn, balance)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    """
    values = (
        row['Receipt No.'],
        row['Completion Time'],
        row['Details'],
        row['Transaction Status'],
        row['Paid In'],
        row['Withdrawn'],
        row['Balance']
    )
    try:
        cursor.execute(sql, values)
    except mysql.connector.Error as err:
        print(f"Error inserting row: {row.to_dict()}")
        print(f"MySQL Error: {err}")
        conn.rollback() # Rollback the transaction on error
        continue # Skip to the next row

conn.commit()
cursor.close()
conn.close()
```

```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Home2020@!",  # change this!
    database="fraud_detection"
)
cursor = conn.cursor()
df = pd.read_csv("cleaned_mpesa.csv")
print(df.columns.tolist())

print("\nNumber of 'nan' values in each column:")
print(df.apply(lambda col: col.astype(str).str.contains('nan').sum()))

# Replace 'nan' with None in the entire DataFrame initially
df = df.replace('nan', None)

import numpy as np

for _, row in df.iterrows():
    sql = """
        INSERT INTO transactions 
        (receipt_no, completion_time, details, transaction_status, paid_in, withdrawn, balance)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
    """
    values = (
        row['Receipt No.'],
        row['Completion Time'],
        row['Details'],
        row['Transaction Status'],
        None if pd.isna(row['Paid In']) else row['Paid In'],
        None if pd.isna(row['Withdrawn']) else row['Withdrawn'],
        None if pd.isna(row['Balance']) else row['Balance'],
    )

    try:
        cursor.execute(sql, values)
    except mysql.connector.Error as err:
        print(f"Error inserting row: {row}")
        print(f"MySQL Error: {err}")


conn.commit()
cursor.close()
conn.close()
```
